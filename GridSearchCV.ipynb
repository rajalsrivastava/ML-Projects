```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species',axis = 1)
Y = df['species']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(X_train,Y_train)
```

```python
model_knn.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(X_train,Y_train)
```

```python
model_svm.score(X_test,Y_test)
```

```python
#now lets use the grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier =GridSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear'],
},cv=5,return_train_score=False)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear'],
},cv=5,return_train_score=False,n_iter=4)
```

```python
classifier_r.fit(X,Y)
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```
